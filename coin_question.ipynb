#Coin flip question

##Question statement:

A coin is fliped, every time that its heads the player gets 100$. However there is a catch, if the player gets tails all the winings are taken away. Moreover if the player gets 3 tails the player is forced to stop playing and all the winings will be taken away from the player

## Optimization to get the most money without loosing:

*here is the equation* : 0.5(0.5(200) + 0.5(87.5)) + 0.5(87.5)

The explinations is as follows, suppose that we are starting the game. We have no tails and are ready to make money. If we get 2 heads in a row meaning if we get 200 we need to leave the game ASAP!Suppose we get one tail then we should flip the coin twice. If we get 2 heads in a row meaning 200$ we should leave the game. Suppose that we have 2 tails, this means if we get tails again we will be out of the game. We need to throw the coin one time, If we get heads once then we need to take the winnings meaning 100 and leave!

## Diffrent personalities and thier behaviours

**The Gambler** - he wants to win as much monmey as possible not taking into account loosing. When there are no tails yet he will not settle for nothing less then 800. If one tail exists he will not settle for nothing less then 600 and if there are 2 tails he will no settle for nothing less then 500.

**The Scardy Cat** - this personality trate is scared to loose. When there are 0 tails then the scrady cat will aim for 200. If there is 1 tail then the scardy cat will be happy with 100 and if there are 2 tails then the scardy cat will pray to get 100



**The Smart Guy** - this personality trait knows the catch and wants to win and not loose. He understands the game there for he aims for 200. When there are 0 tails then he would be happy with 200, if there is 1 tail then 200 is good. Same for 2 tails.



**The Easily Happy** = this is a person that is happy with very little. Meaning the moment that he gets 100 he stops playing
